DATA CLEANING

```python
null_counts=df.isnull().sum()
```

```python
#constructing bar chart for visual representation for null values found in each columns
null_counts.plot(kind='bar', figsize=(10, 6))  
plt.xlabel('columns')
plt.ylabel('no.of null values')
plt.title('Null values in each column')
plt.xticks(rotation=45)
plt.show()
```

```python
null_counts
```

```python
df.columns
```

```python
#filling the null valuesby median method for the column age
df['Age'].fillna(df['Age'].median(),inplace=True)
df
```

```python
#understanding the distribution of values and the common values for the column Embarked.
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna('S', inplace=True)
df
```

```python
df.info()
```

```python
#changing the data types for the columns Survived, PClass, Sex, Age and Embarked.
df['Survived']=df['Survived'].astype('category')
df['Pclass']=df['Pclass'].astype('category')
df['Sex']=df['Sex'].astype('category')
df['Age']=df['Age'].astype('int')
df['Embarked']=df['Embarked'].astype('category')
df.info()
```

```python
sns.countplot(x='Survived', data=df, palette = 'cividis');
plt.show()
```

```python
df.columns
```

```python
df.groupby(['Survived','Sex'])['Survived'].count()
```

```python
#Survival Analysis of males
df[df['Sex'] == 'male'].Survived.groupby(df.Survived).count().plot(kind='pie', figsize=(6, 6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Perished","Survived"])
plt.title("Male survival rate")
plt.show()
```

```python
#Survival Analysis by females
df[df['Sex'] == 'female'].Survived.groupby(df.Survived).count().plot(kind='pie', figsize=(6, 6),explode=[0,0.05],autopct='%1.1f%%')
plt.axis('equal')
plt.legend(["Perished","Survived"])
plt.title("Female survival rate")
plt.show()
```

```python
#Survival Analysis by PClass
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Pclass', data=df , palette = 'summer')
plt.title('Survival Count by PClass')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='PClass', loc='upper right')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
#Age Distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True, color = 'darkviolet')
plt.title('Age Distribution')
```

```python
pd.crosstab([df.Sex,df.Survived],df.Pclass,margins=True).style.background_gradient(cmap='summer_r')
```

```python
print('Oldest person Survived was of:',df['Age'].max())
print('Youngest person Survived was of:',df['Age'].min())
print('Average person Survived was of:',df['Age'].mean())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = 'Embarked', data=df, palette = 'rocket')
plt.title('Embarked Distribution')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.show()
```
